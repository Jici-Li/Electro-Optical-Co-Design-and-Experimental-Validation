1. Introduction 

1.1 Research Background

MZM (Mach-Zehnder modulators) is important in silicon photonic systems. Ideal sin²-response MZMs must have perfect periodicity, but practical devices usually displays non-ideal properties such as period stretching and bias shifting  Traditional method of recording data manually is hard to capture these tiny anomalies.

1.2 Research Method Overview

This study uses both simulation and experimental methods: first analyzing MZM characteristics through VPI simulation, then verifying with Python-controlled real MZM devices. This approach checks simulation accuracy while testing real device performance. By comparing simulation and experimental results, we gain a more complete understanding of MZM behavior, providing reliable references for optical communication system design.

1.3 Contribution

This study develops a Python-based automation tool for VPI simulations to precisely control MZM input voltages. The experiments reveal a sin²-like response (vs. theoretical cos² expectation), traced to VPI's default quadrature bias and hidden parameters (fixed bias of pi/4, leading to a mixure of sin² and cos² for output ). This automated approach provides standardized control for complex photonic device simulations, while exposing parameter mapping discrepancies between commercial tools and theoretical models.

(2) Python Automation

Voltage sweep is implemented by Python with 10ms per step:

```python
import socket
import numpy as np
import time

class MZMController:
    def __init__(self, host='127.0.0.1', port=5465):
        self.sock = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
        self.sock.connect((host, port))
        self._send("INIT MZM Vpi=3.14")  
        assert self._recv() == "READY"
    
    def set_voltage(self, voltage):
        self._send(f"SET VOLTAGE {float(voltage):.4f}")
        return self._recv() == "OK"

    def _send(self, cmd):
        self.sock.sendall((cmd + '\n').encode())

    def _recv(self):
        return self.sock.recv(1024).decode().strip()

    def close(self):
        self.sock.close()

def sweep_voltage():
    mzm = MZMController()
    try:
        voltages = np.linspace(0, 3.14, 10)  
        dt = 0.1  
        
        for v in voltages:
            start_time = time.time()
            mzm.set_voltage(v) 
            elapsed = time.time() - start_time
            if elapsed < dt:
                time.sleep(dt - elapsed)
    finally:
        mzm.close()

if __name__ == "__main__":
    sweep_voltage()
```

(3) Data Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt

power_data = pd.read_csv('D:/Users/lij16/MZM_report/d.csv')
signal_data = pd.read_csv('D:/Users/lij16/MZM_report/data_2.csv')  

merged_data = pd.merge(signal_data, power_data, on='Time(ms)', how='inner')

signal = merged_data['Electrical Signal(a.u.)'].values

start_idx = 0
while signal[start_idx] > 0.1:  
    start_idx += 1

end_idx = start_idx
while end_idx < len(signal) and signal[end_idx] < 3.14:
    end_idx += 1

cropped_data = merged_data.iloc[start_idx:end_idx]

plt.figure(figsize=(10, 6))
plt.plot(cropped_data['Electrical Signal(a.u.)'], cropped_data['Power(mW)'], 'b-', linewidth=2)
plt.xlabel('Voltage (V)', fontsize=12)
plt.ylabel('Power (mW)', fontsize=12)
plt.title('Voltage(0→3.14) vs Power', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
